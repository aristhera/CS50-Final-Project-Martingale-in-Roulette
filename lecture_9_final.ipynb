```python
# Martingale in Roulette

from random import randint
def main():
    x = get_int()
    bet = x[0]
    money = x[1]
    init = bet
    print('\n1.) Doubling your bet\n')
    d = doubling(bet, money, init)
    print(d)
    print('\n2.) Average number of rounds + EV:\n')
    print(proba(d, init))
    print("""\n3.) 50 Tests: See how many rounds it takes to lose + how often < 100 rounds.
    Compare with expected rounds in proba\n""")
    liste = []
    for i in range(1, 51):
        r = roulette(init, bet, money)
        print(i, r)
        if r[0] < 100:
            liste.append(r[0])
    print(f"\nThis makes {len(liste)} rounds < 100 to lose, see here: {liste}")

def get_int():
    while True:
        try:
            b = int(input("Please enter an amount of your bet: "))
            m = int(input("Please enter your total amount of money: "))
            return b, m
        except ValueError:
            print("bet/money is not an integer")

def roulette(i, b, m):  # i = init, b = bet, m = money  
    n = 0
    while m > b:
        z = number()
        if even(z):
            m = m - b
            b = b * 2
        else:
            m += b
            b = i
        n += 1
    return n, z, b, m
    
def doubling(b, m, i):
    n = 0
    while m > b:
        n += 1
        m -= b
        print(n, b, m)
        b *=2
    return n, b-i 

def number():
    return randint(0, 36)

def even(z):
    if z % 2 == 0:
        return True

def proba(d, i): 
    q = (19/37) ** d[0]
    p = 1 / q
    ev = (1 - q) * i - (q * d[1])
    return f"You will probably lose every {round(p, 1)} rounds \n" \
           f"+ the expected value per round is {round(ev, 2)}$ loss. \n" \
           "The roulette-function will show you a great deal of outliers, so you better stop gambling!"
    
if __name__ == "__main__":
    main()
```
